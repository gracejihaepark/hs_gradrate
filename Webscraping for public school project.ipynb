```python
import requests
from bs4 import BeautifulSoup as BS
page = requests.get('https://data.nysed.gov/profile.php?instid=7889678368')
```

```python

soup = BS(page.content, 'html.parser')
```

```python
soup
```

```python
all_districts=soup.find_all('li', class_='bullet-item')
```

```python
all_districts[10]
i=0
districts = []
bad = []
for item in all_districts:
    if i <10:
        bad.append(item)
        i+=1
    elif i == 42:
        bad.append(item)
        i+=1
    else:
        districts.append(item)
        i+=1
```

```python
#clicking one of the districts
page = requests.get('https://data.nysed.gov/profile.php?instid=800000048532')

```

```python
high_schools_list = [a['href'] for a in high_schools]
```

```python
high_schools_list[0].replace('profile.php?instid=','')
```

```python
high_school_ids = []
for item in high_schools_list:
    thing=item.replace('profile.php?instid=','')
    high_school_ids.append(thing)
```
